```python
import numpy as np
import matplotlib.pyplot as plt

size = 22
x = np.arange(size)
total_width, n = 0.8, 4
width = total_width / n
x = x - (total_width - width) / 2

fig, ax = plt.subplots(figsize=(12, 3))
ax.bar(x, np.log(np.array(sdql_exec_time_compiler_thread_1)),  width=width, label='sdql (1-thread)', color="deepskyblue")
ax.bar(x + width, np.log(np.array(sdql_exec_time_compiler_thread_8)),  width=width, label='sdql (8-thread)', color="chocolate")
ax.bar(x + 2 * width, np.log(np.array(duckdb_exec_time)),  width=width, label='duckdb', color="gold")
ax.bar(x + 3 * width, np.log(np.array(pandas_exec_time)), width=width, label='pandas', color="lightseagreen")
ax.set_xlabel('TPC-H Query')
ax.set_ylabel('Execution Time (ms)')
ax.set_xticks(x, x_axis_labels)
ax.set_yticks(np.log(np.array([10, 100, 1000, 10000])), [10, 100, 1000, 10000])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20),
          ncol=4, fancybox=True, shadow=True)
plt.show()
```
